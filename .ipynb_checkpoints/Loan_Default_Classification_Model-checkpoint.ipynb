## Business Understanding

An asset based credit lending company,that offers vehicle loans has contracted us, to predict the rate of loan default, based on their existing customer history.They need to create a better credit risk scoring model, to minimize the default rate to less than 10% of their portfolio.

The dataset is of loans given for motorvehicles. We have been tasked with building an asset loan default prediction, for motorvehicles based on the given dataset. We will be predicting the rate of default using the Logistic Regression, Decision Trees and Random Forest predictive models.
Our target variable,rate of loan default, is  based on KYC(Know Your Customer) data collected on the loan borrowers. This data includes the borrower's personal information, employment status and credit status. 

The 5C's are often used when accessing credit worthiness, which are Capital, Capacity, Collateral, Character and Conditions.


### Overview

Financial institutions incur significant losses due to loan defaults. By modeling loan defaults, these institutions can make better informed decisions.
It can gelp inform 
* Pricing decisions and price in risk
* Risk evaluation of loan
The loan default dataset for motor vehicles is a valuable resource for financial institutions or lenders who offer loans for purchasing motor vehicles. The dataset provides insights into the factors that contribute to loan defaults in the context of motor vehicle financing. By analyzing this dataset, businesses can gain a better understanding of the risks associated with lending for motor vehicles and make informed decisions to minimize potential losses.

### Business Problem

A dataset of vehicle loan issued in India - LNT Financial Services
The problem revolves around predicting and preventing loan defaults for motor vehicle loans. Loan defaults occur when borrowers fail to make loan payments as agreed, resulting in financial losses for the lender. Identifying the factors that contribute to loan defaults can help businesses assess the creditworthiness of applicants and implement appropriate risk mitigation strategies.

### Objectives


1. Predictive Modeling: Develop a machine learning model that accurately predicts the likelihood of loan default for motor vehicle loans based on various borrower and loan-related features.

2. Risk Assessment: Identify the key factors that significantly influence the probability of loan default. This includes analyzing the impact of variables such as credit scores, employment type, loan amount, loan-to-value ratio, borrower's age, etc., on the likelihood of default.

3. Portfolio Management: Segment the loan portfolio based on the predicted default probabilities to allocate appropriate resources for monitoring and managing high-risk loans. This helps in prioritizing collection efforts and implementing proactive measures to mitigate potential defaults.

4. Risk Mitigation Strategies: Based on the insights gained from the analysis, develop strategies to reduce the overall default rate. This may involve refining the loan approval process, setting stricter eligibility criteria, adjusting interest rates, offering targeted financial counseling to borrowers, or exploring insurance options to protect against default losses.

5. Operational Efficiency: Optimize the loan processing and underwriting procedures by identifying bottlenecks, reducing manual efforts, and streamlining the decision-making process. This improves efficiency, reduces costs, and enhances the overall customer experience.

By focusing on these objectives, financial institutions can better assess the creditworthiness of borrowers, minimize loan defaults, and ultimately mitigate financial risks associated with motor vehicle loans.

```python
# Assess the relationship between flag columns and loan default
from scipy.stats import chi2_contingency
flags = df[['Mobileno_avl_flag',
            'Aadhar_flag',
            'Pan_flag',
            'Voterid_flag',
            'Driving_flag', 
            'Passport_flag']]


# Iterate over each categorical feature
for feature in flags:
    # Create a contingency table of the feature and target variable
    contingency_table = pd.crosstab(df[feature], df['Loan_default'])
    
    # Perform chi-square test of independence
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)
    
    # Print the results
    print("Feature:", feature)
    print("Chi-square statistic:", chi2)
    print("P-value:", p_value)
    print()
```

>  Based on the chi-square test results, the 'Aadhar_flag', 'Voterid_flag', and 'Passport_flag' features show strong evidence of a significant relationship with the 'Loan_default' target variable. The 'Driving_flag' feature also shows some evidence of a significant relationship. However, the 'Mobileno_avl_flag' and 'Pan_flag' features do not show significant evidence of a relationship with the 'Loan_default' target variable.

```python
# Based on domain knowlegde, we will drop the flags with low predictive power
df = df.drop(columns=['Mobileno_avl_flag',
                      'Pan_flag',
                      'Driving_flag'], axis=1)
df.shape
```

```python
# H0 - No diffrence in means of the disbursed amount between the loans defaulted and the non-defaults
# H1 - There is a significant difference in means between the disbursed amounts in the loans defaulted and non-defaults
# Parametric Test - Independent t-test

default_group = df[df['Loan_default'] == 1]['Disbursed_amount']
non_default_group = df[df['Loan_default'] == 0]['Disbursed_amount']
t_statistic, p_value = stats.ttest_ind(default_group, non_default_group)
print("Independent t-test results:")
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
```

 A very low p-value suggests strong evidence against the null hypothesis and indicates a significant difference between the groups. the p-value is extremely small (close to zero), indicating a highly significant difference between the default and non-default groups in terms of the 'DISBURSED_AMOUNT' variable.

```python
# Independent ttest

default_group = df[df['Loan_default'] == 1]['Asset_cost']
non_default_group = df[df['Loan_default'] == 0]['Asset_cost']
t_statistic, p_value = stats.ttest_ind(default_group, non_default_group)
print("Independent t-test results:")
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
```

```python
# Filter the DataFrame for the desired categories in 'Employment_type'
categories = ['Salaried', 'Self Employed']
filtered_df = df[df['Employment_type'].isin(categories)]

# Create a contingency table of the filtered data for 'Employment_type' and 'Loan_default'
contingency_table = pd.crosstab(filtered_df['Employment_type'], filtered_df['Loan_default'])

# Perform the chi-square test
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Print the results
print("Chi-square test results:")
print("Chi-square statistic:", chi2)
print("P-value:", p_value)
```

chi-square statistic is 0.0 and the p-value is 1.0, it suggests that there is no significant association between the categories 'Salaried' and 'Self Employed' in the 'Employment_type' variable and the 'Loan_default' variable. In other words, the default rates are not significantly different between these two employment types.

Based on the provided results, some variables have very low p-values (e.g., Disbursed_amount, Ltv, Employment_type), indicating a potentially significant association with loan default. Conversely, variables with higher p-values (e.g., Sec_overdue_accts, Driving_flag) may have weaker associations with loan default.

# Classification Modeling

If capturing all actual loan defaults is more important, prioritize recall. Recall (also known as sensitivity or true positive rate) measures the proportion of correctly predicted positive instances (loan defaults) out of all actual positive instances. Recall focuses on the ability of the model to capture all positive instances, which is crucial in identifying all actual loan defaults and minimizing false negatives.

```python
# Classification modeling libraries
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, f1_score, accuracy_score, recall_score\
,roc_curve, auc, precision_score, RocCurveDisplay
```

```python
# Function for model evaluation 
# Create a an evaluation metrics function, for testing our models

def eval_model(model, X_test, y_test, n_classes=None):
    preds = model.predict(X_test)
    probs = model.predict_proba(X_test)
    
    accuracy = accuracy_score(y_test, preds)
    recall = recall_score(y_test, preds)
    precision = precision_score(y_test, preds)
    f1 = f1_score(y_test, preds)
    con_max = confusion_matrix(y_test, preds)
    
    if n_classes is None:
        n_classes = model.classes_
    
    plt.figure(figsize=(8 ,6))
    disp = ConfusionMatrixDisplay(con_max, display_labels=["default", "not_default"])
    disp.plot(cmap=plt.cm.magma)
    plt.title("Confusion Matrix")
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show();

    print("\n")
    print("Accuracy: ", accuracy)
    print("Precision: ", precision)
    print("Recall: ", recall)
    print("F1: ", f1)
    
    #Show ROC Curve 
    fpr, tpr, threshold = roc_curve(y_test, probs[:,1], pos_label=1)
    roc_auc = auc(fpr, tpr)
    print("AUC: ", roc_auc)

     # Create and display the ROC curve
    roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc)
    roc_display.plot()
    
    # Plot the chance level line
    plt.plot([0, 1], [0, 1], color='darkorange', linestyle='--')
    plt.show();

    results_df = pd.DataFrame()
    results_df['true_class'] = y_test
    results_df['predicted_class'] = list(preds)
    results_df['default_prob'] = probs[:, 1]

    #plot the distribution of probabilities for the estimated classes 
    sns.distplot(results_df[results_df['true_class'] == 0]['default_prob'], label="No Default", hist=False)
    sns.distplot(results_df[results_df['true_class'] == 1]['default_prob'], label="Default", hist=False)
    plt.show()
    
    #see the true class versus predicted class as a percentage
    print(results_df.groupby('true_class')['predicted_class'].value_counts(normalize=True))
```

> Interpreting the evaluation metrics for the logistic classifier:

1. **Accuracy:** The accuracy of the model is approximately 58%. This means that out of all the loans we predicted, about 58% were correctly classified as either defaulting or not defaulting.

2. **Precision:** Precision is around 29%. When the model predicts a loan will default, it's correct about 29% of the time. This indicates that the model might sometimes label loans as defaulting when they actually don't.

3. **Recall (Sensitivity):** The recall is about 65%. This measures how well the model can find loans that are actually defaulting. So, the model identifies about 65% of the actual loan defaults.

4. **F1 Score:** The F1 score is approximately 40%. This is a balanced measure of precision and recall. It takes into account both false positives and false negatives. An F1 score of 40% suggests that the model has a moderate balance between precision and recall.

5. **AUC (Area Under the Curve):** The AUC is around 64%. This is a measure of how well the model distinguishes between defaulting and non-defaulting loans. An AUC of 64% indicates that the model's ability to differentiate between the two classes is decent.

In simple terms, our model does an okay job at identifying loans that may default, but there's room for improvement. It's able to catch some actual default cases, but there are also times when it makes mistakes. We can work on refining the model to make better predictions and help reduce the risk of loan defaults.

When you have a high accuracy but low recall and precision, it suggests that your model is making a significant number of false negatives (missed positive cases) while achieving a high overall accuracy. This situation typically occurs when you have imbalanced class distribution in your dataset, where the majority class dominates and the minority class is underrepresented.
In summary, when you have high accuracy but low recall and precision, it implies that your model is performing well in terms of overall correctness but is struggling to correctly identify positive cases. This is often seen in imbalanced datasets, where the model tends to predict the majority class more frequently, leading to lower recall and precision for the minority class.

Based on the client's requirements, the most important thing was to reduce loan_defaults.
In as much as this is at the expense of the model accuracy, the risk of false negatives is much higher.

- Asset_cost
- Disbursed_amount
- Credit_history_length_months

The loan default classifier demonstrates a moderate level of performance. It correctly classifies around 58.0% of the instances, which indicates that it is able to distinguish between default and non-default cases to some extent.

- **Accuracy:** The model's accuracy is approximately 58.0%, implying that the percentage of correctly predicted outcomes is relatively decent.

- **Precision:** With a precision of about 29.3%, the model's ability to correctly identify true positive cases (default) is limited, which could result in some false positive predictions.

- **Recall:** The recall value is around 64.9%, meaning the model successfully captures a significant portion of actual default cases, though it may still miss some.

- **F1 Score:** The F1 score, a harmonic mean of precision and recall, stands at approximately 40.3%. This suggests that the model achieves a balance between precision and recall, but improvements could be made in both aspects.

- **AUC (Area Under the Curve):** The AUC score is about 64.5%, indicating that the model's ability to rank positive instances higher than negative ones is somewhat satisfactory.

In terms of model training and generalization:

- **Training Accuracy:** The model performs slightly better on the training data with an accuracy of 58.3%. This suggests that the model has learned to capture some patterns present in the training data.

- **Testing Accuracy:** The testing accuracy matches the overall accuracy at approximately 58.0%, indicating that the model's performance on new, unseen data is consistent with its performance on the training data.

Overall, the classifier shows potential but could benefit from further tuning and improvements, especially in terms of precision and recall. Evaluating different strategies like feature engineering, hyperparameter tuning, and considering other algorithms might lead to enhanced performance.
